```python
import pandas as pd
```

```python
filepath = 'KangerDataFile.csv'
```

```python
data = pd.read_csv(filepath)
```

```python
data.info()
```

```python
# plot with pandas
import matplotlib.pyplot as plt
data.plot(subplots=True)
plt.savefig('subplotKanger.jpg')
```

```python
data.plot()
plt.yscale('log')
plt.show()
```

```python
column_list1 = ['BD','LOI']
data[column_list1].plot()
plt.show()
```

```python
column_list1 = ['BD','LOI']
data[column_list1].plot()
plt.yscale('log')
plt.show()
```

```python
ax = data.plot(kind='scatter', x='BD', y='LOI')
ax.set_xlim((0,2))
ax.set_ylim((0,60))
ax.set_xlabel("BD (g/cm3)")
ax.set_ylabel("LOI (%)")
import matplotlib.pyplot as plt
plt.savefig('scatterBDLOI.jpg')
```

```python
groupby_catchment = data.groupby('Cthcm')
groupby_catchment.plot(subplots=True)
```
